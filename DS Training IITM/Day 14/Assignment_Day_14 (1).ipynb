```python
# Internet Required
df_titanic = sns.load_dataset('titanic')
df_titanic.sample()
```

```python
# sns.countplot     #julius_Ai : subsitute for Data Analyst
# sns.barplot
# sns.boxplot
# sns.scatterplot
# sns.catplot
# sns.pairplot
# sns.heatmap
# sns.regplot
# sns.lmplot
# sns.lineplot
# sns.kdeplot
# sns.violinplot
```

```python
import matplotlib.pyplot as plt
sns.countplot(data = df_titanic, x = 'sex', hue = 'alive', palette = sns.color_palette('magma',2)) #hue: further divide based on categorical()
plt.title('Titanic Survival Analysis Based On Gender')
plt.show()
```

```python
# Barplot

sns.barplot(data = df_titanic, x = 'class', y = 'fare', hue = 'alive', estimater = 'mean') #min, max, mean
plt.show()
# Confidence Interval(CI): 95%
```

```python
sns.barplot(data = df_titanic, x = 'class', y = 'fare', hue = 'alive') #min, max, mean
# plt.grid(color = 'r', alpha = '0.1')
plt.show()
```

```python
# sns.barplot

sns.barplot(data = df_titanic, x= 'class', y = 'fare',hue = 'alive', errorbar= ('ci',0)) # min, max, sum
plt.grid(color = 'r',alpha = 0.1)
plt.show()
# Confidence Interval(CI): 95%
```

```python
# # sns.boxplot: To detect Outliers and to check data distribution
# # To detect Outliers:
# # Range: Min to max(19 - 22): 35, 28, 45(Outliers: fliers

# # Q2: Quantile : 50%
# # Q1: 25%
# # Q3: 75%

# # IQR = Q3 - Q1
# UW(Upper cap) = Q3 + 1.5*IQR
# LW(Lower cap) = Q1 - 1.5* IQR
```

```python
# sns.boxplot

sns.boxplot(data = df_titanic, x='class', y   = 'fare' , hue = 'sex')
plt.show()
```

```python
# sns.boxplot
plt.figure(figsize = (10,5))
sns.boxplot(data = df_titanic, y='class', x  = 'fare' , hue = 'sex',orient ='h')
plt.show()
```

```python
# sns.violin Plot
plt.figure(figsize = (10,5))
sns.violinplot(data = df_titanic, x='class', y = 'fare' , hue = 'sex',)
plt.show()
```

```python
# new data_set
df_pen = sns.load_dataset('penguins')
df_pen.sample()
```

```python
sns.violinplot(data = df_pen, x = 'species', y = 'body_mass_g', hue = 'sex')
plt.show()
```

```python
# sns.scatterplot: Numerical vs Numerical Analysis
# corr: 

sns.scatterplot(data = df_pen, x = 'bill_length_mm', y = 'bill_depth_mm', hue = 'species')
plt.show()
```

```python
df_tips = sns.load_dataset('tips')
df_tips.sample()
```

```python
# catplot

sns.catplot(data = df_tips, x = 'time' , y = 'total_bill',kind = 'bar', hue = 'sex', row = 'smoker', col = 'day' ) #kind strip
plt.show()
```

```python
sns.catplot(data = df_tips, x = 'time' , y = 'total_bill',kind = 'violin', hue = 'sex', row = 'smoker', col = 'day' ) #kind strip
plt.show()
```

```python
sns.lmplot(data = df_tips, x = 'total_bill' , y = 'tip')
plt.show() #kind = 'scatter', hue = 'sex', row = 'smoker', col = 'day' ) 
```

```python
sns.lmplot(data = df_tips, x = 'total_bill' , y = 'tip', row = 'smoker', col = 'day' , hue = 'sex')
plt.show()
```

```python
# Relplot
sns.relplot(data = df_tips, x = 'total_bill' , y = 'tip' ,kind = 'scatter', hue = 'sex', row = 'smoker', col = 'day')
plt.show()
```

```python
# Pairplot

df_iris = sns.load_dataset('iris')
df_iris.sample()
```

```python
df_iris.iloc[:,:-1].corr()
```

```python
sns.pairplot(data = df_iris)
plt.show()
```

```python
sns.pairplot(data = df_iris,hue = 'species',corner = True)
plt.show()
```

```python
# kdeplot: kernel density estimation

sns.kdeplot(data = df_iris, x = 'sepal_length', y = 'petal_width', hue = 'species')
sns.scatterplot(data = df_iris, x = 'sepal_length', y = 'petal_width', hue = 'species')
plt.show()
```

```python

sns.kdeplot(data = df_iris, x = 'sepal_length', y = 'sepal_width', hue = 'species')
# sns.scatterplot(data = df_iris, x = 'sepal_length', y = 'sepal_width', hue = 'species')
plt.show()
```

```python

sns.jointplot(data = df_iris, x = 'sepal_length', y = 'sepal_width', hue = 'species')
plt.show()
```

```python
sns.jointplot(data = df_iris, x = 'sepal_length', y = 'sepal_width', kind = 'reg')
plt.show()
```
